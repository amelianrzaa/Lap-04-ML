```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Pada bagian ini, dilakukan import library yang akan digunakan dalam program, yaitu Pandas untuk manipulasi data, NumPy untuk operasi matematika, dan Matplotlib untuk visualisasi data.

## DATAFRAME

```python
data = {'Nama': ['Putri', 'Intan', 'Nahida', 'Amelia'],
        'Tinggi': [159, 164, 155, 160],
        'Berat': [68, 60, 47, 64]}

df = pd.DataFrame(data)
```

Pada bagian ini, dibuat DataFrame yang berisi data tinggi dan berat badan dari empat orang.

## Pembuatan Model

```python
#Model Data Uji
from sklearn.model_selection import train_test_split
x = df['Tinggi'].values.reshape(-1,1)
y = df['Berat'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#Model Data Regresi Linear
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

Pada bagian ini, dilakukan pembuatan model regresi linear dengan menggunakan library Scikit-Learn. Data dibagi menjadi data latih dan data uji dengan perbandingan 70:30. Kemudian, model regresi linear dibuat dan dilatih dengan data latih.

## Prediksi pada Data Uji

```python
y_pred = model.predict(x_test)
print(f'Koefisien Regresi: {model.coef_[0]}')
```

Pada bagian ini, dilakukan prediksi pada data uji dan ditampilkan koefisien regresi.

## Visualisasi pada DataFrame

```python
plt.scatter(x, y, color='blue')
plt.plot(x, y, color='red', linewidth=3)
plt.title('Data Tinggi & Berat Badan')
plt.xlabel('Tinggi (cm)')
plt.ylabel('Berat Badan (kg)')
plt.show()
```

Pada bagian ini, dilakukan visualisasi data tinggi dan berat badan pada DataFrame dengan scatter plot dan garis regresi linear.

```python
tinggi_prediksi = [[160]]
berat_prediksi = model.predict(tinggi_prediksi)
print(f'Prediksi Berat Badan untuk Tinggi 160 cm : {berat_prediksi[0]}')
```

```python
berat_prediksi = [[64]]
tinggi_prediksi = model.predict(berat_prediksi)
print(f'Prediksi Tinggi Badan untuk Berat 64 kg : {tinggi_prediksi[0]}')
```

Pada bagian ini, dilakukan prediksi tinggi/bobot badan dengan menggunakan model yang telah dibuat sebelumnya. Prediksi dilakukan dengan memasukkan nilai tinggi/bobot badan yang ingin diprediksi ke dalam model.
